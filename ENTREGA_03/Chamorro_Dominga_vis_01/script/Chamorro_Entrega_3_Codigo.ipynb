```python
# Group by Year and Gender and count occurrences
gender_counts = df.groupby(['Ano', 'Genero']).size().reset_index(name='Count')

# Calculate the total count per year
total_counts_per_year = df.groupby('Ano').size().reset_index(name='Total')

# Merge the counts with the total counts
gender_percentage = pd.merge(gender_counts, total_counts_per_year, on='Ano')

# Calculate the percentage
gender_percentage['Percentage'] = (gender_percentage['Count'] / gender_percentage['Total']) * 100

# Pivot the table to have Year as index, Gender as columns, and Percentage as values
gender_percentage_table = gender_percentage.pivot(index='Ano', columns='Genero', values='Percentage').fillna(0)

display(gender_percentage_table)
```

```python
# Calculate the average percentages for Femenino and Masculino across all years
average_female_percentage = gender_percentage_table['Femenino'].mean()
average_masculino_percentage = gender_percentage_table['Masculino'].mean()

# Create a new row for the average percentages
average_row = pd.DataFrame({'Femenino': [average_female_percentage], 'Masculino': [average_masculino_percentage]}, index=['Average'])

# Remove the Gender_Gap column from the original table if it exists
if 'Gender_Gap' in gender_percentage_table.columns:
    gender_percentage_table = gender_percentage_table.drop(columns=['Gender_Gap'])

# Append the average row to the table
gender_percentage_table_with_average = pd.concat([gender_percentage_table, average_row])

# Remove the Gender_Gap column from the final table
if 'Gender_Gap' in gender_percentage_table_with_average.columns:
    gender_percentage_table_with_average = gender_percentage_table_with_average.drop(columns=['Gender_Gap'])

display(gender_percentage_table_with_average)
```

```python
import altair as alt

# Reset index to make 'Ano' a column
gender_percentage_long = gender_percentage_table_with_average.reset_index()

# Remove the 'Average' row
gender_percentage_long = gender_percentage_long[gender_percentage_long['index'] != 'Average']

# Melt the DataFrame to long format
gender_percentage_long = gender_percentage_long.melt(id_vars='index', var_name='Genero', value_name='Percentage')

# Rename the 'index' column to 'Ano'
gender_percentage_long = gender_percentage_long.rename(columns={'index': 'Ano'})

# Convert 'Ano' to integer type
gender_percentage_long['Ano'] = gender_percentage_long['Ano'].astype(int)

# Create the stacked area chart
chart = alt.Chart(gender_percentage_long).mark_area().encode(
    x=alt.X('Ano:O', title='Ao'),  # Use ordinal for discrete years
    y=alt.Y('Percentage', title='Porcentaje'),
    color='Genero',
    tooltip=['Ano', 'Genero', 'Percentage']
).properties(
    title='Porcentaje de gnero por ao'
)

chart.display()
```
